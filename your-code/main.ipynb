#### 4. Import the users table 

```python
user_tab = pd.read_sql_query('SELECT * FROM stats.users', engine)
```

#### 5. Rename Id column to userId

```python
user_tab.rename(columns = {'Id': 'userId'}, inplace = True)
```

#### 6. Import the posts table. 

```python
post_tab = pd.read_sql_query('SELECT * FROM stats.posts', engine)
```

#### 7. Rename Id column to postId and OwnerUserId to userId

```python
post_tab.rename(columns = {'Id': 'postId', 'OwnerUserId': 'userId'}, inplace = True)
```

#### 8. Define new dataframes for users and posts with the following selected columns:
    **users columns**: userId, Reputation,Views,UpVotes,DownVotes
    **posts columns**: postId, Score,userID,ViewCount,CommentCount

```python
user_df = user_tab[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]
post_df = post_tab[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
merge_df = pd.merge(right=user_df, left=post_df, on='userId')
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = merge_df.isnull().sum()
null_cols
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
merge_df['ViewCount'].isnull().mean().round(4)
cols_eliminate = list(null_cols[null_cols >= 40].index)
data_clean = merge_df.drop(cols_eliminate, axis=1)
```

#### 11. Adjust the data types in order to avoid future issues. Which ones should be changed? 

```python
data_clean['userId'] = pd.to_numeric(data_clean['userId'], downcast='integer')
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
for columna in data_clean._get_numeric_data():
    print(columna)
    print(data_clean[columna].plot.box())
    plt.show()
    print('--------' * 10)
```

```python
stats = data_clean.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats.head(5)
```

```python
outliers = pd.DataFrame(columns=data_clean.columns)

for col_name in stats.index:
    iqr = stats.loc[col_name, 'IQR']
    corte = iqr * 1.5
    corte_bajo = stats.loc[col_name, '25%'] - corte
    corte_alto = stats.loc[col_name, '75%'] + corte
    resultado = data_clean[(data_clean[col_name] < corte_bajo) | 
                           (data_clean[col_name] > corte_alto)]
    resultado['Atipico'] = col_name
    print(resultado)
    print()
    outliers = outliers.append(resultado)
```

```python
import os  
os.makedirs('your-code', exist_ok=True)  
outliers.to_csv('your-code/outliers.csv')
```
